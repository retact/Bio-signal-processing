```python
%cd /content/drive/My Drive/Biological_signal/
%matplotlib inline
import cv2
from skimage.io import imread
import matplotlib.pylab as plt
```

##**Task1**

```python
img = cv2.imread("./Bio-signal-processing/lecture9/sample_image/IMG_6205.png")
hsv= cv2.cvtColor(img, cv2.COLOR_BGR2HSV) 
hsv[..., 1] = 100
hsv[..., 2] = 200
img_hsv= cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.imshow(img_hsv)
```

##**Task2**


```python
img_gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

threshold = 128
ret, img_th= cv2.threshold(img_gray, threshold, 255, cv2.THRESH_BINARY)
hist = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
ret, img_otsu= cv2.threshold(img_gray, 0, 255, cv2.THRESH_OTSU)

plt.figure(figsize=(6, 3))
plt.subplots_adjust(wspace=0.4, hspace=0.6)
plt.subplot(2, 2, 1)
plt.title('input')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.title('grayscale image')
plt.imshow(img_gray, cmap='gray')
plt.subplot(2, 2, 3)
plt.title('binary image')
plt.imshow(img_th, cmap= "gray")
plt.subplot(2, 2, 4)
plt.title('Otsu-binary image')
plt.imshow(img_otsu, cmap= "gray")
plt.show()

```
