## **TASK1**


```python
import cv2
import numpy as np
import matplotlib.pylab as plt
img= cv2.imread("src/Lena.jpg")

#　Average Filter
average_kernel = np.ones((9,9)) / 81.0
img_ave = cv2.filter2D(img, -1, average_kernel)
# Gaussian Filter
img_ga = cv2.GaussianBlur(img, (9, 9), 2)
# Median Filter
img_me = cv2.medianBlur(img,9)
# Bilateral Filter
img_bi= cv2.bilateralFilter(img, 20, 30, 30)

# plot
plt.figure(figsize=(12, 12))
plt.subplots_adjust(wspace=0.2, hspace=0.3)
plt.subplot(2, 2, 1)
plt.title('Average')
plt.imshow(cv2.cvtColor(img_ave, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.title('Gaussian')
plt.imshow(cv2.cvtColor(img_ga, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 3)
plt.title('Median')
plt.imshow(cv2.cvtColor(img_me, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 4)
plt.title('Bilateral')
plt.imshow(cv2.cvtColor(img_bi, cv2.COLOR_BGR2RGB))
plt.show()
```

## **TASK2**

```python
img_ga = cv2.GaussianBlur(img, (3, 3), 2)
# 標準偏差を変更
img_ga2 = cv2.GaussianBlur(img, (3, 3), 10)
# カーネルサイズを変更
img_ga3 = cv2.GaussianBlur(img, (9, 9), 2)
# どちらも変更
img_ga4 = cv2.GaussianBlur(img, (9, 9), 10)

# plot
plt.figure(figsize=(12, 12))
plt.subplots_adjust(wspace=0.2, hspace=0.3)
plt.subplot(2, 2, 1)
plt.title('basic')
plt.imshow(cv2.cvtColor(img_ga, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.title('SD only')
plt.imshow(cv2.cvtColor(img_ga2, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 3)
plt.title('size only')
plt.imshow(cv2.cvtColor(img_ga3, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 4)
plt.title('SD and size')
plt.imshow(cv2.cvtColor(img_ga4, cv2.COLOR_BGR2RGB))
plt.show()
```

まず,カーネルサイズを大きくすることで全体のぼかしがより強くなることがわかる．これはより広いカーネルの大きさにより平滑化を行っているため，広い範囲の影響を受けることによる作用だと考えるおkとができる．次に標準偏差を広げることでカーネルサイズ内での中心部の影響が小さくなることから標準偏差をあげることでより広い範囲の平滑化すなわち，座標的にいうより高い数値をパスしてしまうことからSDとsizeの4つ目の画像ではより強いフィルタがかかっているように見えると考えることができる．
