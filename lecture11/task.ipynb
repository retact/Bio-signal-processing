## **TASK1**


```python
import cv2
import numpy as np
import matplotlib.pylab as plt
img= cv2.imread("src/Lena.jpg")

#ã€€Average Filter
average_kernel = np.ones((9,9)) / 81.0
img_ave = cv2.filter2D(img, -1, average_kernel)
# Gaussian Filter
img_ga = cv2.GaussianBlur(img, (9, 9), 2)
# Median Filter
img_me = cv2.medianBlur(img,9)
# Bilateral Filter
img_bi= cv2.bilateralFilter(img, 20, 30, 30)

# plot
plt.figure(figsize=(12, 12))
plt.subplots_adjust(wspace=0.2, hspace=0.3)
plt.subplot(2, 2, 1)
plt.title('Average')
plt.imshow(cv2.cvtColor(img_ave, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 2)
plt.title('Gaussian')
plt.imshow(cv2.cvtColor(img_ga, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 3)
plt.title('Median')
plt.imshow(cv2.cvtColor(img_me, cv2.COLOR_BGR2RGB))
plt.subplot(2, 2, 4)
plt.title('Bilateral')
plt.imshow(cv2.cvtColor(img_bi, cv2.COLOR_BGR2RGB))
plt.show()
```
